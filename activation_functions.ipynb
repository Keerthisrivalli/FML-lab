```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
def tanh(x):
  return np.tanh(x)
```

```python
def relu(x):
  a=[]
  for i in x:
    if i<0:
      a.append(0)
    else:
      a.append(i)
  return a
```

```python
x=np.linspace(-50,50,100)
x
```

```python
plt.plot(x,sigmoid(x))
plt.show()
```

```python
plt.plot(x,tanh(x))
plt.show()
```

```python
plt.plot(x,relu(x))
plt.show()
```
